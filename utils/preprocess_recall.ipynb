```python
from preprocessor import preprocessor
import pandas as pd

df = pd.read_csv('../data/data.csv')
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
pp = preprocessor()
df = pp.process_na(df)
df = pp.classify_columns(df)
```

```python
serial_vc = df['serial'].value_counts()
serial_g2 = serial_vc[serial_vc > 1].index
df_g2 = df[df['serial'].isin(serial_g2)]
```

```python
s = df_g2['mos'].apply(lambda x: x.split(' '))
df_g2['mos'] = s
display(df_g2)
```

```python
df_g2 = df_g2.sort_values(by=['serial', 'call_day','call_time'])
recall_counts = df_g2.groupby(['serial','call_day'])['call_time'].apply(lambda x: x.shape[0])
```

```python
df_rc = pd.merge(df_g2, recall_counts, on=['serial','call_day'], how='left')
display(df_rc)
```
