```python
import pickle
import os

def load_models():
    models = []
    modelnames = []
    for file in os.listdir('results/models'):
        if file.endswith('.pkl'):
            with open('results/models/' + file, 'rb') as f:
                models.append(pickle.load(f))
                modelnames.append(file)
    return models, modelnames


models, modelnames = load_models()
```

```python
selected_models = []
for modelname, model in zip(modelnames, models):
    if modelname.startswith('XGBoost_drop'):
        selected_models.append((modelname, model))
models = [model for (name, model) in selected_models[-5:]]
bestmodel = models[-1]
```

```python
import xgboost as xgb
from matplotlib import pyplot as plt

# ... (Assume you have a trained XGBoost model named 'model')

xgb.plot_importance(bestmodel)
plt.show()
```
